```python
import pandas as pd
import numpy as np


df1 = pd.read_csv("..\datasets\cyprus_airport_air_traffic\csv\LARNAKA AIRPORT DAILY AIR TRAFFIC_2018-2019 TOTAL.csv")
df2 = pd.read_csv("..\datasets\cyprus_airport_air_traffic\csv\LARNAKA AIRPORT DAILY AIR TRAFFIC_2020-2021 TOTAL.csv")
df = pd.concat([df1,df2])
df['City'] = 'Larnaka'
df['station_code'] = 5

df3 = pd.read_csv("..\datasets\cyprus_airport_air_traffic\csv\PAFOS AIRPORT DAILY AIR TRAFFIC_2018-2021 TOTAL.csv")
df3['City'] = 'Pafos'
df3['station_code'] = 15

df = pd.concat([df,df3])
print(df.shape)
df.head()

df.to_csv('CYPRUS_AIRPORT_DAILY_AIR_TRAFFIC_2018_2021.csv', header=True)
```
